```python
#analysing the distribution of aggregate ratings
plt.hist(df['Aggregate rating'],edgecolor='Black')
plt.xlabel('Aggregate ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate ratings ')
plt.show()
```

```python
#determining the most common rating range:
rating_counts=df['Aggregate rating'].value_counts()
most_common_rating=rating_counts.idxmax()
count_of_most_common_rating=rating_counts.max()
print('Most common rating is: ',most_common_rating,'with',count_of_most_common_rating,'as occurances')
```

```python
#Calculating the average number of votes received by restaurants:
avg_votes=df['Votes'].mean()
print('average number of votes received by restaurants are: ',avg_votes)
```
